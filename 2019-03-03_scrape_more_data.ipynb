```python
df = pd.read_csv('./data/dataset.csv')
```

```python
DOMAIN = 'https://www.federalreserve.gov'
```

```python
result = []
```

```python
for link in df.l[len(result):]:
    try:
        target = DOMAIN + link
        print(target)
        r = requests.get(target)
        soup = BeautifulSoup(r.text)
        text = soup.select_one('div#article').select('div:nth-of-type(3)')[0].get_text()
        result.append({
            'link': link,
            'content': text
        })
    except:
        print('error', link)
```

```python
pd.DataFrame(result).to_csv('./data/content.csv', index=False)
```
